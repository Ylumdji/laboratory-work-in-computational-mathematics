```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Function1 = lambda x: np.sin(x * x)
Function2 = lambda x: np.cos(np.sin(x))
Function3 = lambda x: np.exp(np.sin(np.cos(x)))
Function4 = lambda x: np.log(x + 3.0)
Function5 = lambda x: np.sqrt(x + 3)
```

```python
def Function1_der(x):
    return 2.0 * x * np.cos(x * x)
    
def Function2_der(x):
    return -np.cos(x) * np.sin(np.sin(x))
    
def Function3_der(x):
    return -np.sin(x) * np.cos(np.cos(x)) * np.exp(np.sin(np.cos(x)))
    
def Function4_der(x):
    return 1.0 / (x + 3.0)
    
def Function5_der(x):
    return 1.0 / (2.0 * np.sqrt(x + 3.0))
```

```python
def method_1(f, x, h):
    return (f(x + h) - f(x)) / h

def method_2(f, x, h):
    return (f(x) - f(x - h)) / h

def method_3(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def method_4(f, x, h):
    return (2 * (f(x + h) - f(x - h)) / (3 * h)) - ((f(x + 2*h) - f(x - 2*h)) / (12 * h))

def method_5(f, x, h):
    return (3 * (f(x + h) - f(x - h)) / (4 * h)) - 3 * ((f(x + 2*h) - f(x - 2*h)) / (20 * h)) + ((f(x + 3*h) - f(x - 3*h)) / (60 * h))
```

```python
h = np.geomspace(1, np.power(2.0, -21), num=22, dtype=float)
x_0 = 26
```

```python

    plt.figure (figsize = (21, 9), dpi = 110)
    plt.title("sin{x^2)", fontsize=30)

    plt.xscale("log")
    plt.yscale("log")

    plt.xlabel('log(h)', fontsize = 24)
    plt.ylabel("log(der. error)", fontsize = 24)

    derivative = Function1_der(x_0)

    errors = []
    for step in h:
        errors.append(np.abs(method_1(Function1, x_0, step) - derivative))
    plt.scatter (h, errors, s = 15, color ="red" , label="method_1")
    plt.plot (h, errors, color="red", linewidth = 1)\

    errors.clear()
    for step in h:
        errors.append(np.abs(method_2(Function1, x_0, step) - derivative))
    plt.scatter (h, errors, s = 15, color ="orange" , label="method_2")
    plt.plot (h, errors, color="orange", linewidth = 1)

    errors.clear()
    for step in h:
        errors.append(np.abs(method_3(Function1, x_0, step) - derivative))
    plt.scatter (h, errors, s = 15, color ="blue" , label="method_3")
    plt.plot (h, errors, color="blue", linewidth = 1)

    errors.clear()
    for step in h:
        errors.append(np.abs(method_4(Function1, x_0, step) - derivative))
    plt.scatter (h, errors, s = 15, color ="yellow" , label="method_4")
    plt.plot (h, errors, color="yellow", linewidth = 1)

    errors.clear()

    for step in h:
        errors.append(np.abs(method_5(Function1, x_0, step) - derivative))
    plt.scatter (h, errors, s = 15, color ="green" , label="method_5")
    plt.plot (h, errors, color="green", linewidth = 1)
    errors.clear()
    plt.legend (loc = "best", fontsize = 14)


    plt.show()
```
